```
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
sns.set(style='darkgrid')

%matplotlib inline
```

```
# 데이터셋
tips = sns.load_dataset('tips')
```

```
tips
```

```
sns.relplot(x='total_bill', y = 'tip', data = tips)
```

```
# hue 색상 - 범주형
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker')
```

```
# style -  hue 색상 - 수치형
sns.relplot(x='total_bill', y = 'tip', data = tips, style = 'smoker', hue = 'size')
```

참고 사이트 : https://seaborn.pydata.org/tutorial/color_palettes.html

```
# 색상
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'size', palette='Blues')
```

```
df = pd.DataFrame(dict(time=np.arange(500), value=np.random.randn(500).cumsum()))
```

```
df
```

```
sns.relplot(x = 'time', y = 'value', data = df, kind = 'line')
```

```
fmri = sns.load_dataset('fmri')
```

```
fmri
```

```
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, ci = None)
```

```
# ci 가 없을 경우 -> 표준편차

sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri)
```

```
# ci = 'sd' 분산

sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, ci = 'sd')
```

```
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, ci = 'sd', hue = 'region', style = 'event')
```

```
# cat plot
sns.catplot(x = 'day', y = 'total_bill', data = tips) # 요일 별 요금 정산
```

```
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'box')
```

```
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'violin')
```

```
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'boxen')
```

```
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'count')
```

```
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'bar')
```

```
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'point')
```

```
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'swarm', hue = 'sex')
```

```
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'violin', inner = 'stick', split = True,  hue = 'sex')
```
